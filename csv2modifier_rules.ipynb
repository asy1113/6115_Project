## Read csv file and change to modifiler_rules

```python
import csv
row_all=[]
with open('KB/new_modifiers_040618.csv') as csvfile:
    csv_file = csv.reader(csvfile, delimiter=',')
    for row in csv_file:
        #print(row[0]+ " --- " +row[1])
        row_all.append(row)
print(len(row_all))

rules1=[]
for row in row_all:
    rules1.append("Comments: ''")
    rules1.append("Direction: "+ row[2])
    rules1.append("Lex: "+ row[0])
    rules1.append("Regex: ''")
    rules1.append("Type: "+ row[1])
    rules1.append("---")

print(len(rules1))
rules='\n'.join(rules1)
print(rules)
        
with open('KB/Part_new_modifiers_040618.yml', 'w') as f:
    f.write(rules)
```
